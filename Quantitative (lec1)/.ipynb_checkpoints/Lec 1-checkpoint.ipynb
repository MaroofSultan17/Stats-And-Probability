```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Data.csv")
data.head()
print("Original Data")
print(data)
```

```python
categories = ['Y', 'N']
user = []
choice = int(input("How many peoples are taken for observation"))
print("would the Chat Gpt is usefull 'Y' or 'N'.")
for x in range(0,choice):
    input_ = input().lower()
    user.append(input_)
print(user)
freq_y = user.count('y')
freq_n = user.count('n')
freq = [freq_y, freq_n]
print(f'Frequency of Y is {freq_y} and Frequency of N is {freq_n}')
no_of_obs = len(user) #rf (Relative frequency = fy / n)
rf_of_fy = (freq_y / no_of_obs)*100
rf_of_fn = (freq_n / no_of_obs)*100
rf = [rf_of_fy, rf_of_fn]
```

```python
# file mn colum bna k is trh add krty hain
disct = {'Catageories': categories, 'Freq': freq, 'Relative Freq': rf}
df = pd.DataFrame(disct)
catg = df['Catageories'].head(2)
f = df['Freq'].head(2)
#graph ka size set krny k lye 
# fig = plt.figure(figsize = (10,7))
plt.title("Lecture 1")
plt.bar(catg, f)
plt.show()
print(df)
```

```python
###Pie chrt for Relative frequency'
#label show krny k lye
my_label = ["YES", "NO"]
# chart mn sy ik specific part ko alehda krny k lye
my_explode = [0.1, 0]
# chart mn colors dyny k lye
my_colors = ["Grey", "Orange"]
# chart ko tittle dyny k lye
plt.title("Relative frequency")
plt.pie(rf, labels = my_label, explode = my_explode, colors = my_colors)
plt.show()
#data ko file mn store krany k lye
df.to_csv('Data.csv')
```
